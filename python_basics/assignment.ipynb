```python
#create a list
my_list=[12,13,14,15,16]
print("My list is:",my_list)
```

```python
#convert it into tuple
my_tuple=tuple(my_list)
print("my tuple is:",my_tuple)
```

```python
#modify a list
my_list[1]=190
print('modifed list:',my_list)
```

```python
#modify a tuple
my_tuple(0)==100
print("modified tuple",my_tuple) 
#this will give error because we cannot modify tuple as it is immutable
```

```python
my_list = [1, 2, 2, 3, 3, 4]
my_set = set(my_list)
print(my_set)
#set removes the duplicate values and only give unique value
```

```python
# Original list
my_list = [10, 20, 30, 40, 50]

# Reverse the list using slicing
reversed_list = my_list[::-1]

print("Original List:", my_list)
print("Reversed List:", reversed_list)
```

```python
#Create a dictionary of student details
student = {
    "name": "Nisha",
    "age": 22,
    "course": "Data Engineering"
}
#Access the name
print("Student Name:", student["name"])


```

```python
#Remove age using pop()
student.pop("age")
```

```python

# Step 4: Print length of dictionary
print("Length of Dictionary:", len(student))
```
