```python
#public acess modifier no restriction ,direct acess allowed
class Employee:
    def __init__(self,name ,salary):
        self.name=name #publis access
        self.salary=salary
emp=Employee("nisha",20000)
print(emp.name)
print(emp.salary)
        
```

```python
#protected acess modifier -python doesnot strictly restricted its warning for developers
class Employee:
    def __init__(self,name,salary):
        self._salary=salary   #protected
emp=Employee("nisha",10000)
print(emp._salary)
    
```

```python
#private access modifier -strong restriction(name mangling)
#python changes name internally -_Employee__salary
class Employee:
    def __init__(self,name,salary):
        self.__salary=salary   #private
emp=Employee("nisha",10000)
# print(emp.__salary) -it will give you attribute error
print(emp._Employee__salary)
```

```python
# Modifying Private Data (Setter Method)
 
class employee:
    def __init__(self, salary):
        self.__salary = salary # 10000 - python does name mangling for private variables- emp._employee__salary = 10000
 
    #Setter Method - update salary safely
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary # 20000
 
    #Getter Method - used to access private data safely
    def get_salary(self):
        return self.__salary
 
 
emp = employee(10000)
# employee.__init(emp,10000)
 
emp.set_salary(20000) # internally - employee.set_salary(emp,20000)
print(emp.get_salary()) # employee.get_salary(emp)
#print(emp.__salary)
 
 
# Flow 1- salary = 10000 -> salary updated too 20000 -> output - 20000
```

Inheritance

```python
class Person:
    def __init__(self,name):
        self.name=name
    def show_name(self):
        print("Name: ", self.name)
 
class Employee(Person): #Inheritaning Person
    def __init__(self, name, salary):
        self.name=name
        self.salary=salary
 
emp=Employee("Nisha", 10000)
emp.show_name() # Inherited method
 
print(emp.salary)
```

```python
class Person:
    def __init__(self,name):
        self.name=name
    def show_name(self):
        print("Name: " ,self.name)
        
class Employee(Person):    #inheriting Person
    def __init__(self, name, salary):
       super().__init__(name)  #calling parent constructor
       #it allows you to call methods from a parent class without explicity naming it
       self.salary=salary
#super call method -it is used to call methods of parent class-to initiliaze parent property ,to reuse to parent logic

emp= Employee("Nisha",10000)
 #inherited method
print(emp.name)
print(emp.salary)

```

Key difference without vs with super
 without super()- parent not initialized,Missing data,Error prone
 with super()- parent initialized,Complete Object,Correct approach

 inheritance-reuse code from parent
 super()-caling parent's setup

 

```python
# create class vehicle:speed
#create a class car:fuel type
#using inheritance
class Vehicle:
    def __init__(self, speed):
        self.speed = speed

    def show_speed(self):
        print("Speed:", self.speed)


# Child class (inherits from Vehicle)
class Car(Vehicle):
    def __init__(self, speed, fuel_type):
        # call parent constructor
        super().__init__(speed)
        self.fuel_type = fuel_type

    def show_details(self):
        print("Speed:", self.speed)
        print("Fuel Type:", self.fuel_type)


# Create object of Car
my_car = Car(120, "Petrol")

# Call methods
my_car.show_details()
```

Polymorphism


```python
#method overriding-when child class provides its own implementation of a method already defined in parent class
class Parent:
    def show(self):
        print("this is a parent")

class Employee(Parent):
    def show(self):  #overriding parent method
        super().show() #call parent method :parent logic +child logic
        print("this is a employee")
obj=Employee()
obj.show()
#python checks first ->child class then->parent class-child method overrides parent method
```

```python
#method overloading-same func name with different number/type of argument -default argument *args
#same function handles diff input
def add (a, b = 0, c = 0):
    return a+b+c
 
#same function behaves differentl based on inputs
print(add(10))
print(add(10,20))
print(add(10,20,30))
```

```python
# *args
 
def add(*numbers):
    return sum(numbers)
 
print(add(10))
print(add(10,20))
print(add(10,20,30))
```
