```python
#constructor(init)
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
s1=Person("sujana",20)
#automaticaaly called everytime the class is being used to create a new object
print(s1.name)
print(s1.age)
```

```python
# using __init__- you set initial values when creating the object
 
class Person2:
    def __init__(self, name, age): # self - pointer - which we will initialize to the objects which we are referring
        self.name = name
        self.age = age
 
p3 = Person2("Nick", 19)
 
print(p3.name)
print(p3.age)
```

```python
#without __init__ - you need to set properties manually for each object
class Person1:
    pass
 
p1 = Person1()
p1.name = "Nisha"
p1.age = 18
 
print(p1.name)
print(p1.age)
```
