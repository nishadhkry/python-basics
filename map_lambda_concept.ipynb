```python
#reccursion 
#function that call itself
#base case-condn to stop the recursion, recursive case-function calling itself

#print numbers 1 to 5
def print_numbers(n):
    if n==6:  #base case
        return
    print(n)
    print_numbers(n+1)  #this is the recursive call

print_numbers(1)
    
        
```

```python
#fibonacci series


#sequence where[0,1,1,2,3,5,8,13,..]

#f(n)=f(n-1)+f(n-2)

```

```python
def fib(n):
    if n<=1: #base case 1
        return n
    else:
        return fib(n-1)+fib(n-2)

print (fib(8))
#f(1)=1 ,f(0)=0,f(2)=1,f(3)=2,f(4)=3,f(5)=5
        

```

```python
 
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
 
#print Series
 
n = 7
for i in range(n):
    print(fib(i))
```

```python
#factorial=5!=5*4*3*2*1
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
